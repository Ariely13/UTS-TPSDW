```python
#install library beautifulsoup4
!pip install requests beautifulsoup4

#kemudian mengimport library
import requests
from bs4 import BeautifulSoup
import json

#list untuk menyimpan data
data_list = []

#membuat looping dari halaman 1-5
for page in range(1, 6):
    url = f"https://loginmegastore.com/search?query=kulkas&p=true&page={page}"
    response = requests.get(url)
    soup = BeautifulSoup(response.text, "html.parser")

#ambil semua produk
    titles = soup.find_all("p", class_="mantine-focus-auto")
    prices = soup.find_all("p", class_="mantine-focus-auto")
    images = soup.find_all("img", class_="mantine-Image-root")

#membentuk loop tiap item
    for i in range(len(images)):
        try:
            title = titles[i].text.strip()
            price = prices[i].text.strip()
            image = images[i]["src"]

            data = {
                "title": title,
                "price": price,
                "image": image
            }

            data_list.append(data)

        except:
            continue

#simpan ke JSON
with open("List_Elektronik.json", "w") as f:
    json.dump(data_list, f, indent=4)
```
